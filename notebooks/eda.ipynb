```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../data/model_training_data.csv')
data
```

```python
plt.figure(figsize=(9,6))
sns.countplot(x='AgeClass', data=data)
# Most of the data is concentrated between the 18-19 and 35-39 age ranges
```

```python
sns.countplot(x='Sex', data=data)
```

```python
sns.histplot(data=data['TotalKg'])
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='BodyweightKg')
```

```python
sns.displot(data=data, x='Age', y='TotalKg')
```

```python
sns.relplot(data=data, x='Age', y='TotalKg', hue='Sex', col='Sex')
```
